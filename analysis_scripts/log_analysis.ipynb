```python
import os
import pandas as pd
from IPython.display import display
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
policy = next(os.walk('./log'))[1]

r = [0.001, 0.01, 0.05, 0.1, 0.2, 0.3, 0.4]
df = pd.DataFrame()
for p in policy:
    path = './log/'+p
    logs = next(os.walk(path))[2]
    for ratio, l in zip(r,logs):
        one_df = pd.read_csv(path+'/'+l)
        one_df['cache_ratio'] = ratio
        df = pd.concat([df, one_df], axis=0)        
```

```python
df
```

```python
f, ax = plt.subplots(figsize=(10, 5))
ax.set(xscale="log")
sns.lineplot(ax=ax, data=df, x='cache_ratio', y='byte_hits/bytes', hue="cacheType")
```
